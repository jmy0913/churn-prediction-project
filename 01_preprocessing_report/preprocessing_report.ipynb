````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from matplotlib import font_manager as fm, rc
import matplotlib.font_manager as fm

font_path = "C:/Windows/Fonts/malgun.ttf"
font_prop = fm.FontProperties(fname=font_path)
rc('font', family=font_prop.get_name())

# ÌïúÍ∏Ä Ìè∞Ìä∏ ÏÑ§Ï†ï
plt.rc('font', family='Malgun Gothic')

# Îç∞Ïù¥ÌÑ∞ Î°úÎìú
df = pd.read_csv('data/E_Commerce_Dataset_model2.csv')

# Ïù¥ÌÉà Í≥†Í∞ù ÏàòÏôÄ ÎπÑÏú® Í≥ÑÏÇ∞
total_customers = len(df)
churned = df['Churn'].sum()
retained = total_customers - churned
churn_rate = (churned / total_customers) * 100
retention_rate = (retained / total_customers) * 100

# Îç∞Ïù¥ÌÑ∞ÏôÄ Î†àÏù¥Î∏î Ï§ÄÎπÑ
sizes = [retained, churned]
labels = [f'Ïú†ÏßÄ Í≥†Í∞ù\n({retained:,}Î™Ö, {retention_rate:.1f}%)',
          f'Ïù¥ÌÉà Í≥†Í∞ù\n({churned:,}Î™Ö, {churn_rate:.1f}%)']

plt.pie(sizes,
        labels=labels,
        autopct='%1.1f%%',
        startangle=90,
        shadow=True,
        textprops={'fontsize': 12})

plt.title('Í≥†Í∞ù Ïù¥ÌÉà ÌòÑÌô© ÏÉÅÏÑ∏ Î∂ÑÏÑù', pad=20, size=15)

# Î≤îÎ°Ä Ï∂îÍ∞Ä
plt.legend(labels,
          title="Í≥†Í∞ù Íµ¨Î∂Ñ",
          loc="center left",
          bbox_to_anchor=(1, 0, 0.5, 1))

plt.tight_layout()
plt.show()

# ÏÉÅÏÑ∏ ÌÜµÍ≥Ñ Ï∂úÎ†•
print("\nÍ≥†Í∞ù Ïù¥ÌÉà ÏÉÅÏÑ∏ ÌÜµÍ≥Ñ:")
print(f"Ï†ÑÏ≤¥ Í≥†Í∞ù Ïàò: {total_customers:,}Î™Ö")
print(f"Ïú†ÏßÄ Í≥†Í∞ù Ïàò: {retained:,}Î™Ö ({retention_rate:.1f}%)")
print(f"Ïù¥ÌÉà Í≥†Í∞ù Ïàò: {churned:,}Î™Ö ({churn_rate:.1f}%)")
````

````python
df.info()
df.describe(include='all')
df.isnull().sum()

df['PreferredLoginDevice'] = df['PreferredLoginDevice'].replace({'Mobile Phone': 'Mobile'})

# ÏàòÏπòÌòï - ÌèâÍ∑†Í∞í ÎåÄÏ≤¥
num_cols = df.select_dtypes(include=['float64', 'int64']).columns.drop('CustomerID')
df[num_cols] = df[num_cols].fillna(df[num_cols].mean())

#Î≤îÏ£ºÌòï - ÏµúÎπàÍ∞í ÎåÄÏ≤¥
cat_cols = df.select_dtypes(include='object').columns
for col in cat_cols:
    df[col] = df[col].fillna(df[col].mode()[0])

# CustomerID Ïª¨Îüº ÏÇ≠Ï†ú
df = df.drop(columns=['CustomerID'])
# Churn - Î≤îÏ£ºÌòïÏúºÎ°ú Î≥ÄÍ≤Ω
df['Churn'] = df['Churn'].astype('category')

df
````

````python
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import seaborn as sns

# ÎÇòÎàîÍ≥†Îîï ÎòêÎäî Í∏∞Î≥∏ ÌïúÍ∏Ä Ìè∞Ìä∏ Í≤ΩÎ°ú ÏßÄÏ†ï
plt.rcParams['font.family'] = 'Malgun Gothic'  # WindowsÏù∏ Í≤ΩÏö∞
plt.rcParams['axes.unicode_minus'] = False  # ÎßàÏù¥ÎÑàÏä§ Íπ®Ïßê Î∞©ÏßÄ



# ÎßåÏ°±ÎèÑ Í∑∏Î£π ÎßåÎì§Í∏∞
def satisfaction_group(score):
    if score <= 2:
        return 'Low (1-2)'
    elif score == 3:
        return 'Medium (3)'
    else:
        return 'High (4-5)'

df['SatisfactionGroup'] = df['SatisfactionScore'].apply(satisfaction_group)

# ÏãúÍ∞ÅÌôî
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='SatisfactionGroup', hue='Churn')
plt.title('ÎßåÏ°±ÎèÑ Í∑∏Î£πÎ≥Ñ Ïù¥ÌÉàÏûê Ïàò')
plt.xlabel('ÎßåÏ°±ÎèÑ Í∑∏Î£π')
plt.ylabel('Í≥†Í∞ù Ïàò')
plt.legend(title='Churn (1=Ïù¥ÌÉà)')
plt.tight_layout()
plt.show()
````

````python
# Í∑∏Î£πÎ≥Ñ Í≥†Í∞ù Ïàò Í≥ÑÏÇ∞
grouped = df.groupby(['SatisfactionGroup', 'Churn']).size().reset_index(name='count')
total_by_group = grouped.groupby('SatisfactionGroup')['count'].transform('sum')
grouped['ratio'] = grouped['count'] / total_by_group

# ÎπÑÏú® ÎßâÎåÄÍ∑∏ÎûòÌîÑ
plt.figure(figsize=(8, 5))
sns.barplot(data=grouped, x='SatisfactionGroup', y='ratio', hue='Churn')
plt.title('ÎßåÏ°±ÎèÑ Í∑∏Î£πÎ≥Ñ Ïù¥ÌÉà ÎπÑÏú®')
plt.xlabel('ÎßåÏ°±ÎèÑ Í∑∏Î£π')
plt.ylabel('ÎπÑÏú®')
plt.legend(title='Churn (1=Ïù¥ÌÉà)')
plt.tight_layout()
plt.show()
````
